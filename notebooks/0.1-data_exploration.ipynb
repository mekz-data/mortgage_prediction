```python
df = df.drop(columns = cols_to_drop)
df.head()
```

```python
drop_rows_cols = ['Original Combined Loan to Value Ratio (CLTV)', 'Number of Borrowers', 'Debt-To-Income (DTI)',  
                  'First Time Home Buyer Indicator', 'Zip Code Short', 'Prepayment Penalty Indicator', 
                  'Interest Only Loan Indicator', 'Borrower Credit Score at Origination']

df = df.dropna(subset=drop_rows_cols)
```

```python
missing_values_count = df.isnull().sum()

# Calculate the total number of rows
total_rows = len(df)

# Calculate the percentage of missing values
missing_percentage = (missing_values_count / total_rows * 100)

# Compute the summary table and display the results
missing_df = missing_percentage.compute()
```

```python
with pd.option_context('display.max_rows', None):
    print(missing_df[missing_df > 0])
```

```python
sample_df = df.sample(frac=0.01)
```

```python
#if Mortgage Insurance Percentage = NaN, and Original Loan to Value Ratio (LTV) <= 80, then  MIP = 0

mip_condition = (sample_df['Mortgage Insurance Percentage'].isna()) & (sample_df['Original Loan to Value Ratio (LTV)'] <= 80)
```

```python
sample_df['Mortgage Insurance Percentage'] = sample_df['Mortgage Insurance Percentage'].mask(mip_condition, 0)
sample_df = sample_df.persist()
```

```python
result = sample_df.comput()
result.head()
```
